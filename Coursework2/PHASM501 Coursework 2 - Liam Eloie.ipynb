# Krylov Subspace Solvers

There are many ways for solving a matrix equation $Ax = b$. When working with more complicated systems where direct solvers become unable to solve the system efficiently, one naturally turns to iterative solvers -- providing approximate solutions to these matrix equations. Amongst the most popularly used iterative solvers for sparse matrices are the Krylov subspace solvers. 

Consider the iteration

$$x_{k+1} = (I - A)x_k + b$$

This iterative equation, if to converge, will converge to the solution of $Ax = b$. Starting with $x_1 = b$, it is possible to show that the $k^{th}$ iteration will lead to a solution that can be expressed in terms of a polynomial in $A$

$$x_k = P_{k-1}(A)b \approx A^{-1}b$$

where $P_{k-1}(A) = a_0b + a_1 Ab + ... + a_{k-1}A^{k-1}b$
In order for $x_k$ to be an approximate solution to $A^{-1}b$, it is desired to choose the cooefficients for $P_{k-1}(A)b$ such that 

$$P_{k-1}(A)b \approx A^{-1}b$$

To go about choosing $P_{k-1}(A)b$, the $k^{th}$ Krylov Subspace $\mathcal{K}_k(A, b)$ is defined

$$\mathcal{K}_k(A, b) = \text{span}\{b, Ab, ..., A^{k-1}b\}$$

Additionally, the corresponding residual is also defined

$$r_k = b - Ax_k$$

Different Krylov subspace methods are defined depending on what kind of condition is given to the residual.

* GMRES/ MINRES - $r_k \perp \mathcal{K}_k(A, b)$
* Conjugate Gradients - $\min_{\mathcal{K}_k(A, b)}\{\perp r_k\}$

## Krylov Subspace Solvers - How does the rate of convergence change when the mesh size h changes?



GMRES, MINRES, and CG, although come form very similar origins, possess very different convergence behaviour. To get a general overview of the behaviour of scalability for each solver, the plots on the left-hand side demonstrates how the number of iterations increases as the grid dimension increases for all different types of diffusion: constnat, smooth, rough, and anisotropic. It is possible to infer that the general trend for each solver doesn't necessarily depend on the diffusion and that the following can be said about the solvers:

* GMRES - The number of iterations required to solve the system grows exponentially with the mesh size, h.
* CG - The number of iterations required to solve the system grows linearly with the mesh size, h.
* MINRES - The number of iterations required to solve the system grows linearly with mesh size, h, but with a declined gradient compared to CG.

On the right hand side, plots for the residual convergence (log10) for each solver as the mesh dimension, h, increases can be seen (with a constant diffusion). 

* GMRES - The convergence is linear, in log10, for a given grid dimension. The spacing between the ends of the convergenece grow exponentially -- this is another way of showing the exponential relationship between the number of iterations vs the mesh size, h.
* MINRES/ CG - The convergence is non-linear -- both solvers converge more slowly at the start as the grid dimension, h, increases, but quickly make up for it to provide a linear spacing between the ends of the covergence lines.

An interesting aside is to note that GMRES converges almost extactly to the convergence criteria and doesn't vary as grid dimension increases, MINRES tends to stop *before* the convergence criteria and increases as dimension size increases, and CG tends to stop *after* the convergence criteria and decreases as dimension size increases.

MINRES is the best performing solver for all types of diffusion, with CG coming in second place, and GMRES coming in last.

```python
%matplotlib inline
from system import System
from solver import Solver
from compare_solvers import CompareSolvers

from dolfin import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python

diffusion_types = [
                   {'type': 'constant', 'expression': Constant(1.0)}, 
                   {'type': 'smooth', 'expression': Expression('x[0]', degree=1)},
                   {'type': 'rough', 'expression': None},
                   {'type': 'anisotropic', 'expression': Expression((("0.01", "0"), ("0", "100")), degree=2)}
                ]

solvers = [
    {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': None},
    {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': None},
    {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': None},
]

dimensions = [i for i in range(20, 101, 10)]

for solver in solvers:
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3.5))
    for diffusion_type in diffusion_types:
        compare_solvers = CompareSolvers(solvers, diffusion_type['expression'], diffusion_type['type'])
        iterations = compare_solvers.generate_iterations(dimensions)
        ax1.plot(dimensions, iterations[solver['solver_type']], label=diffusion_type['type'])
        
        if(diffusion_type['type'] == 'constant'):
            residuals = [(dimension, compare_solvers.generate_residuals(dimension)) for dimension in dimensions]
            colors = cm.rainbow(np.linspace(0, 1, len(residuals)))
            for (index, residual_tuple), c in zip(enumerate(residuals), colors):
                dim, res_dict = residual_tuple

                ax2.plot(np.log10(res_dict[solver['solver_type']]), color=c, label=dim)

            ax2.set_xlabel('Number of iterations')
            ax2.set_ylabel('Residual (log10)')
            ax2.legend(loc=1, bbox_to_anchor=(1.4, 1), title="Grid Dimension")
        
    ax1.set_xlabel('Grid dimension')
    ax1.set_ylabel('Number of iterations')
    ax1.set_title(solver['solver_type'].upper())
    ax1.legend(loc=2, bbox_to_anchor=(1, 0.98), title="Diffusion type")
    
    plt.tight_layout(pad=0.4, w_pad=20, h_pad=1.0)
    plt.show()
```

## What happens for different diffusion functions such as constant diffusion, smoothly varying diffusion or rough diffusion?

Below are the residual plots (log10) for three Krylov solvers: gmres, minres, and cg. For each solver, three different diffusion types were chosen: constant, smooth, and rough. Constant diffusion problems were solved in the least amount of iterations for all solvers, which is to be expected due to the symmetry of the problem. Interestingly, the performance of the solvers on smooth and rough diffusion are similar. This was unexpected as the system is more complicated with rough diffusion.

## Compare Krylov subspace solvers with algebraic multigrid

Algebraic multigrid methods can also be used to solve a matrix equation. Below, I considered three different types of AMG solvers: Ruge-Stuben, Rootnode, and Smoothed Aggregation. The first immediate difference in all these residual plots, compared to Krylov solvers, is that there is clear a decrease in the number of iterations required to solve the system. They're all fairly linear and clustered closer together. This close clustering suggests that these methods have great scalability when it comes to larger systems. It can be seen, that in general, Ruge-Stuben outperforms Smoothed Aggregation, which in turn outperforms Rootnode.

```python
diffusion_types = [
                   {'type': 'constant', 'expression': Constant(1.0)}, 
                   {'type': 'smooth', 'expression': Expression('x[0]', degree=1)},
                   {'type': 'rough', 'expression': None}
                ]

solvers = [
    {'solver_type':'amg', 'subsolver_type': 'ruge_stuben', 'precondition_type': None},
    {'solver_type':'amg', 'subsolver_type': 'smoothed_aggregation', 'precondition_type': None},
    {'solver_type':'amg', 'subsolver_type': 'rootnode', 'precondition_type': None},
]

dimensions = [i for i in range(20, 101, 10)]

for solver in solvers:
    f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 3.5))
    ax_array = [ax1, ax2, ax3]
    for dif_index, diffusion_type in enumerate(diffusion_types):
        compare_solvers = CompareSolvers(solvers, diffusion_type['expression'], diffusion_type['type'])

        residuals = [(dimension, compare_solvers.generate_residuals(dimension)) for dimension in dimensions]
        colors = cm.rainbow(np.linspace(0, 1, len(residuals)))

        for (index, residual_tuple), c in zip(enumerate(residuals), colors):
            dim, res_dict = residual_tuple

            ax_array[dif_index].plot(np.log10(res_dict[str(solver['solver_type']) + ": " + str(solver['subsolver_type'])]), color=c, label=dim)

        ax_array[dif_index].set_xlabel('Number of iterations')
        ax_array[dif_index].set_ylabel('Residual (log10)')
        ax_array[dif_index].set_title(solver['solver_type'] + ": " + solver['subsolver_type'] + " - " + diffusion_type['type'])
        ax_array[dif_index].legend(loc=1, bbox_to_anchor=(1.4, 1), title="Grid Dimension")
    
    plt.tight_layout(pad=0.4, w_pad=20, h_pad=1.0)
    plt.show()
```

## Test AMG cycles as preconditioners for a Krylov method and compare to other preconditioners

The AMG preconditioners used for the Krylov solvers were: Rube-Stuben, Smoothed Aggregation, and Rootnode. These were compared with each other, as well as with ILU.

```python
def krylov_with_preconditions(dimension):
    diffusion_types = [
                       {'type': 'constant', 'expression': Constant(1.0)},
                    ]

    solver_list = [
        [{'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'ilu'},
        {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}],
        [{'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'ilu'},
        {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}],
        [{'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'ilu'},
        {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}]
    ]

    dimensions = [i for i in range(20, dimension + 1, 1)]

    for sol_index, solvers in enumerate(solver_list):
        for diffusion in diffusion_types:
            compare_solvers = CompareSolvers(solvers, diffusion['expression'], diffusion['type'])
            residuals = [(dimension, compare_solvers.generate_residuals(dimension)) for dimension in dimensions]

            legend_counter = 0
            for res in residuals:
                dim, res_list = res

                res_ilu = res_list[solvers[0]['solver_type'] + ": " + solvers[0]['precondition_type']]
                res_smoothed = res_list[solvers[1]['solver_type'] + ": " + solvers[1]['precondition_type']]
                res_root = res_list[solvers[2]['solver_type'] + ": " + solvers[2]['precondition_type']]
                res_ruge = res_list[solvers[3]['solver_type'] + ": " + solvers[3]['precondition_type']]

                plt.plot(np.log10(res_ilu), color='r', label=solvers[0]['precondition_type'] if legend_counter == 0 else "")
                plt.plot(np.log10(res_smoothed), color='c', label=solvers[1]['precondition_type'] if legend_counter == 0 else "")
                plt.plot(np.log10(res_root), color='b', label=solvers[2]['precondition_type'] if legend_counter == 0 else "")
                plt.plot(np.log10(res_ruge), color='m', label=solvers[3]['precondition_type'] if legend_counter == 0 else "")
                legend_counter += 1
            plt.title(solvers[0]['solver_type'].upper()) 
        plt.title("Convergence graphs for " + solvers[0]['solver_type'].upper() + " with preconditioners \n for grid dimensions between 20 and " + str(dimension))
        plt.legend(loc=1, bbox_to_anchor=(1.8, 1), title="Preconditioners")
        plt.show()
```

Considering only constant diffusion, it was found that for lower dimensional problems (below 78), the Krylov solvers preconditioned with all these methods behave well, and have a significant decrease in the amount of iterations needed to solve the equation. Specifically, the ranking for best preconditioners are as follows:

* ILU
* Ruge-Stuben
* Smoothed Aggregation
* Rootnode

These rankings extend to smooth and rough diffusions.

```python
krylov_with_preconditions(40)
```

It is when we get to larger grids that the ILU preconditioner starts to misbehave and have a significant increase in the amount of iterations needed to solve the equation. For the example of constant diffusion, it can be seen that on the 80th dimension, there is a large increase in iterations -- especially in the case of MINRES and CG. The solvers converge, but not to the specified residual of $10^{-5}$. This problem extends to other the other diffusion types (smooth, rough).

```python
krylov_with_preconditions(80)
```

By removing the ILU preconditioner from the comparison, it can be seen from the plot below that the AMG preconditioners are very robust and scalable. 

```python
def krylov_with_preconditions_without_ilu(dimension):
    diffusion_types = [
                       {'type': 'constant', 'expression': Constant(1.0)},
                    ]

    solver_list = [
        [{'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}],
        [{'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}],
        [{'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'smoothed_aggregation'},
        {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'rootnode'},
        {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': 'ruge_stuben'}]
    ]

    dimensions = [i for i in range(100, dimension + 1, 1)]

    for sol_index, solvers in enumerate(solver_list):
        for diffusion in diffusion_types:
            compare_solvers = CompareSolvers(solvers, diffusion['expression'], diffusion['type'])
            residuals = [(dimension, compare_solvers.generate_residuals(dimension)) for dimension in dimensions]

            legend_counter = 0
            for res in residuals:
                dim, res_list = res

                res_smoothed = res_list[solvers[0]['solver_type'] + ": " + solvers[0]['precondition_type']]
                res_root = res_list[solvers[1]['solver_type'] + ": " + solvers[1]['precondition_type']]
                res_ruge = res_list[solvers[2]['solver_type'] + ": " + solvers[2]['precondition_type']]

                plt.plot(np.log10(res_smoothed), color='c', label=solvers[0]['precondition_type'] if legend_counter == 0 else "")
                plt.plot(np.log10(res_root), color='b', label=solvers[1]['precondition_type'] if legend_counter == 0 else "")
                plt.plot(np.log10(res_ruge), color='m', label=solvers[2]['precondition_type'] if legend_counter == 0 else "")
                legend_counter += 1
            plt.title(solvers[0]['solver_type'].upper()) 
        plt.title("Convergence graphs for " + solvers[0]['solver_type'].upper() + " with preconditioners \n for grid dimensions between 100 and " + str(dimension))
        plt.legend(loc=1, bbox_to_anchor=(1.8, 1), title="Preconditioners")
        plt.show()
```

## What happens if you use strongly anisotropic diffusion?

The plot below demonstrates what happens to each of the anisotropic solvers when you introduce an anisotropic diffusion. As the anisotropic diffusion constant increases, each of the solvers take a shorter amount of iterations to converge. There is a large decrease in the amount of the iterations required for convergence for larger anisotropic diffusion when the grid dimension is below 40 -- after that, there is a large jump.

```python
def anisotropic_dict(epsilon):
    epsilon_recip = 1./epsilon
    return {'type': 'anisotropic', 'expression': Expression(((str(epsilon_recip), "0"), ("0", str(epsilon))), degree=2)}

diffusion_types = [
                       {"dict": anisotropic_dict(1), "ani": 1},
                       {"dict": anisotropic_dict(10), "ani": 10},
                       {"dict": anisotropic_dict(100), "ani": 100},
                       {"dict": anisotropic_dict(1000), "ani": 1000},
                       {"dict": anisotropic_dict(10000), "ani": 10000},
                       {"dict": anisotropic_dict(100000), "ani": 100000},
                    ]

solvers = [
    {'solver_type':'minres', 'subsolver_type': None, 'precondition_type': None},
    {'solver_type':'gmres', 'subsolver_type': None, 'precondition_type': None},
    {'solver_type':'cg', 'subsolver_type': None, 'precondition_type': None},
]

dimensions = [i for i in range(20, 61, 5)]

for solver in solvers:
    for diffusion_type in diffusion_types:
        compare_solvers = CompareSolvers(solvers, diffusion_type["dict"]['expression'], diffusion_type["dict"]['type'])
        iterations = compare_solvers.generate_iterations(dimensions)
        
        if(solver['solver_type'] == 'amg'):
            key = solver['solver_type'] + ": " + solver['subsolver_type']
            print key
        else:
            key = solver['solver_type']

        plt.plot(dimensions, iterations[key], label=diffusion_type['ani'])
        
    plt.xlabel('Grid dimension')
    plt.ylabel('Number of iterations')
    plt.title(solver['solver_type'].upper())
    plt.legend(loc=2, bbox_to_anchor=(1, 0.98), title="Anisotropic Constant")
    
    plt.tight_layout(pad=0.4, w_pad=20, h_pad=1.0)
    plt.show()
```
